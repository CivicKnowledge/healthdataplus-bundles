```python
%matplotlib inline
from ambry import get_library
l = get_library()
b = l.bundle('census.gov-acs_p5ye2014-poverty-hdp-0.0.1')
w = b.warehouse('acs_counties')
#w.clean()
w.dsn
```

```python
w.query("""

MATERIALIZE census.gov-acs-p5ye2014-c18131;
MATERIALIZE census.gov-acs-p5ye2014-c27016;
MATERIALIZE census.gov-acs-p5ye2014-b17020;

""").close()
```

```python
w.query("""

-- Get only counties in California

CREATE TABLE counties AS 
SELECT gvid, name AS county_name FROM census.gov-tiger-2015-counties
WHERE statefp = 6;

INDEX counties (gvid);


""").close()
```

```python
df = w.dataframe("""
SELECT * FROM counties AS idx
LEFT JOIN census.gov-acs-p5ye2014-b17020 AS b17020 ON idx.gvid = b17020.gvid
LEFT JOIN census.gov-acs-p5ye2014-c18131 AS c18131 ON idx.gvid = c18131.gvid
LEFT JOIN census.gov-acs-p5ye2014-c27016 AS c27016 ON idx.gvid = c27016.gvid
;
""")
```
